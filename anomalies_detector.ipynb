```python
import pandas as pd
import pickle
from tqdm import tqdm
from datetime import datetime
from sklearn.metrics import classification_report
```

```python
with open('metrics.pickle', 'rb') as handle:
    unserialized_metrics = pickle.load(handle)

with open('logs.pickle', 'rb') as handle:
    unserialized_logs = pickle.load(handle)
metrics = []
first = True
for data in tqdm(unserialized_metrics):
    if first:
        metrics = pd.DataFrame(data)
        first = False
    else:
        metrics = pd.concat([metrics, pd.DataFrame(data)], ignore_index=True)
metrics.to_csv('metrics.csv')
logs = []
first = True
for data in tqdm(unserialized_logs):
    if first:
        logs = pd.DataFrame(data)
        first = False
    else:
        logs = pd.concat([logs, pd.DataFrame(data)], ignore_index=True)
logs.to_csv('logs.csv')
```

```python
metrics = pd.read_csv('metrics.csv').drop(['Unnamed: 0'],axis=1)
metrics
```

```python
logs = pd.read_csv('logs.csv', index_col='timestamp', parse_dates=["timestamp"]).drop(['Unnamed: 0'],axis=1)
metrics.index= [datetime.fromtimestamp(x) for x in logs.index]
metrics.sort_index(inplace=True)
metrics.head(10)
```

```python
def id_attack(df=metrics):
    metrics.index= [datetime.fromtimestamp(x) for x in logs.index]
    metrics.sort_index(inplace=True)
    kek = metrics.groupby("user_id").rolling('5s').value.count()
    aa = kek[kek > 50].reset_index()
    return sorted(aa.user_id.unique())
```

```python
print(classification_report(sorted(metrics[metrics.attack == 1].user_id.unique()), id_attack()))
```
